```python
# Función para hacer un value_counts de tods las categorias de la base de datos

def value_counts_all_columns(df):
    value_counts_dict = {}  # Diccionario para almacenar los resultados de value_counts

    for column in df.columns:
        value_counts = df[column].value_counts()
        value_counts_dict[column] = value_counts

    return value_counts_dict
```
