```python
# Librerias generales
import pandas as pd
import numpy as np
import re
# Librerias e visualización
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Lectura de datos cronicos
df_cronicos = pd.read_csv('DBs\df_cronicos_listo.csv')
df_cronicos.drop(['Unnamed: 0','atencion','nro_ingreso', 'mes','year'], axis=1, inplace=True)
df_cronicos = df_cronicos.rename(columns={'year': 'year_enfermedad', 'mes': 'mes_enfermedad'})
df_cronicos
```

```python
# Leer base de datos egresos
df_egresos = pd.read_csv('DBs\df_egresos.csv')
df_egresos.drop(['Unnamed: 0','nro_atencion','mes','year'], axis=1, inplace=True)
df_egresos = df_egresos.rename(columns={'year': 'year_egreso', 'mes': 'mes_egreso'})
df_egresos
```

```python
# Leer base de datos usuarios
df_usuarios = pd.read_csv('DBs\df_usuarios.csv')
df_usuarios.drop('Unnamed: 0', axis=1, inplace=True)
df_usuarios = df_usuarios.rename(columns={'year': 'year_activo_programa', 'mes': 'mes_activo_programa'})
df_usuarios
```

```python
# Se borraron las variables siguientes porque se tenía información parecida en las otras bases de datos
df_usuarios=df_usuarios.drop(['quinquenio','ultima_clase_funcional','primera_clase_funcional','year_activo_programa','mes_activo_programa'],axis=1) 
```

```python
# Se ordenaron los datos por la edad de mayor a menor
df_usuarios=df_usuarios.sort_values(by='edad', ascending=False)
```

```python
df_usuarios = df_usuarios.drop_duplicates(subset=['nrodoc']) 
```

```python
df_usuarios
```

```python
# Realizar el primer "left join" entre df_egresos y df_usuarios
resultado_temp = pd.merge(df_egresos , df_cronicos, on='nrodoc', how='inner')
```

```python
resultado_temp=resultado_temp.dropna()
```

```python
# Realizar el segundo "left join" entre resultado_temp y df_cronicos
resultado_final = pd.merge(resultado_temp ,  df_usuarios, on='nrodoc', how='inner')
```

```python
pd.set_option('display.max_columns', None)
resultado_final
```

```python
# se filtro para asegurarnos que todos lo pacientes estuvieran bajo la modalidad de pago global prospectivo 
categoria_deseada = 'PGP'
df_filtrado = resultado_final[resultado_final['modalidad_contrato'] == categoria_deseada] 
```

```python
# Se identifica la existencia de muchas filas duplicadas, por lo cual se decide eliminar
filas_duplicadas = resultado_final.duplicated(keep='first')
cantidad_duplicadas = filas_duplicadas.sum()
cantidad_duplicadas
```

```python
df_filtrado = resultado_final.drop_duplicates()
```

Se optó por aplicar un filtro en la base de datos para eliminar registros con números de documento (nrodoc) e ingresos idénticos, con el objetivo de mantener registros de estadía únicos. Esta estrategia se implementó para asegurar la precisión en la predicción del tiempo de permanencia de los pacientes en el hospital.

La presencia de números de documento duplicados o tiempos de estancia iguales en la base de datos podría introducir un sesgo en el modelo de predicción y distorsionar los resultados. La duplicación de registros podría llevar a una sobreponderación de ciertos datos y, en consecuencia, a una falta de representatividad en el modelo.

```python
#se ordena el dataframe para seleccionar el mes en que ingresaron 
df_filtrado= df_filtrado.sort_values(by='fecha_ingreso_clinica')
```

```python
# Se eliminan las filas duplicadas
df_filtrado = df_filtrado.drop_duplicates(subset=['nrodoc', 'nro_ingreso']) 
```

```python
# Se convierte a formato datetime y se excluyen las fechas anteriores a 2021
df_filtrado['fecha_ingreso_clinica'] = pd.to_datetime(df_filtrado['fecha_ingreso_clinica'])
df_filtrado = df_filtrado[df_filtrado['fecha_ingreso_clinica'].dt.year >= 2021]
```

```python
# Solo se tuvieron en cuenta los servicios mayores o iguales a 24 horas
df_filtrado = df_filtrado[df_filtrado['servicio_habilitado'] != 'Urgencias < 24 horas'] 
```

```python
#Se borro debido a que era una variable numerica y tenia muchos nulos dificiles de reemplazar
df_filtrado.drop('cuantos_cigarrillos_dia',axis=1, inplace=True)
```

* Borrar 'microalbuminuria', 'constantes' 

Se ha decidido eliminar las columnas 'microalbuminuria' y 'constantes' de nuestro conjunto de datos. Estas columnas se caracterizan por su falta de variabilidad, lo que significa que no contribuyen de manera significativa a la información del modelo. 

* Borrar 'nro_ingreso', 'modalidad_contrato'

Se decide eliminar estas dos variables ya que el nro_ingreso es un codigo y la modalidad_contrato tiene todos sus datos PGP
* Convertir categoricas: 'auto-calificacion_nivel_de_ejercicio': 

Vamos a transformar la columna 'auto-calificacion_nivel_de_ejercicio' en una variable categórica, ya que esta columna representa una calificación otorgada por el paciente a su nivel de ejercicio y se comporta de manera categórica en nuestros análisis.


```python
df_filtrado.drop(['microalbuminuria', 'constantes','nro_ingreso','modalidad_contrato','fecha_ingreso_clinica'],axis=1, inplace=True)
```
